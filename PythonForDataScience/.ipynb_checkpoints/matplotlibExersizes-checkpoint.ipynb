```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax.plot(x,y)
ax2.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_title('zoom')
ax.plot(x,z)
ax2.plot(x,z)
ax2.set_xlim([20,22])

plt.show()
```

```python
fig,ax = plt.subplots(1,2)
plt.show()
```

```python
fig,ax = plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show()
```

```python
fig,ax = plt.subplots(1,2, figsize=(12,2))
ax[0].plot(x,y, color = 'red', ls= '--',marker='+', markersize=5)
ax[1].plot(x,z)
plt.show()
```
