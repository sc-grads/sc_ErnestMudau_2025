 You create objects (instances of classes)
python
Copy code
fig = plt.figure()
plt.figure() returns a Figure object (an instance of Matplotlib’s Figure class).

python
Copy code
axes = fig.add_axes([0.1,0.1,0.8,0.8])
fig.add_axes(...) returns an Axes object (an instance of the Axes class).

These are actual Python objects, not just functions printing a chart.

2. You call methods on those objects
python
Copy code
axes.plot(x, y)
plot() is a method of the Axes object.

python
Copy code
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
These are all methods that change the state of the Axes object (its labels, title, etc.).

3. Objects store state
The Axes object stores information about the data you plotted, the labels, and the title.

The Figure object stores the overall canvas, including the Axes objects, size, background, etc.

4. You tell the object to display itself
python
Copy code
plt.show()
This tells Matplotlib’s backend to render the Figure object with all its current state.

In short:
This is OOP because you’re:

Creating objects (Figure, Axes)

Calling methods on those objects to modify them

Using encapsulated state inside those objects

Interacting with class instances rather than just calling procedural functions.
